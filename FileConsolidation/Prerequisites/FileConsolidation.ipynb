Function to get Download Time

```python
def getDownloadTime(path):
    c_time = os.path.getctime(path)
    downloadTime = time.ctime(c_time)
    return downloadTime
```

Function to get Latest File

```python
def getLatestFile(path,fileType = 'csv'): 
    
    path+='/*.' + fileType
    list_of_files = glob.glob(path) 
    latest_file = max(list_of_files, key=os.path.getctime)
    return latest_file
```
