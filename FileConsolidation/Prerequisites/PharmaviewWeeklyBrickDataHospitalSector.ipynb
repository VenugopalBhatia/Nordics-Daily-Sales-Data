```python
driver = webdriver.Chrome(chromeDriver,options = options)
driver.maximize_window()
driver.get('https://dynamics.pharmaview.dk/ords/f?p=300:101')

driver.find_element_by_id("P101_USERNAME").send_keys(username)
driver.find_element_by_id("P101_PASSWORD").send_keys(password)
driver.find_element_by_id('P101_LOGIN').click()
timeout = 500
initialCount = len(os.listdir(defaultDirectory))
newCount = initialCount
retries = 0

while(initialCount == newCount and retries<5):
    
    try:
        try:
            timeoutMenuBar = 100
            locatorMainMenu = '//*[@id="direct_menu_menubar_6i"]'
            WebDriverWait(driver, timeoutMenuBar).until(EC.visibility_of_element_located((By.XPATH, locatorMainMenu)))
        except:
            driver.refresh()
            timeoutMenuBar = 30
            locatorMainMenu = '//*[@id="direct_menu_menubar_6i"]'
            WebDriverWait(driver, timeoutMenuBar).until(EC.visibility_of_element_located((By.XPATH, locatorMainMenu)))
#         
        try:
#             print("yes")
            driver.find_element_by_link_text('Weekly Brick Data').click()
        except:
#             print("wbd exception")
#             driver.find_elements_by_xpath('/html/body/form/div[1]/div[1]/div[2]/main/div[2]/div/div[1]/div/div/div/div/div/div[1]/ul/li')[-1].click()
            driver.find_element_by_css_selector("#direct_menu_menubar_6i").click()
            time.sleep(3)
#             driver.find_element_by_link_text('Weekly Brick Data').click()
        timeout3 = 100
        locatorWeeklyBrick = '//*[@id="data_sheet_weekly_heading"]'  
        WebDriverWait(driver, timeout3).until(EC.visibility_of_element_located((By.XPATH, locatorWeeklyBrick)))
#         Need to add a wait that ensures weekly brick data has been clicked
        time.sleep(2)
        timeout2 = 500
        locator2 = '//*[@id="show_filters"]'
        WebDriverWait(driver, timeout2).until(EC.visibility_of_element_located((By.XPATH, locator2)))
        try:
            time.sleep(5)
            driver.find_element_by_id('show_filters').click()
        except:
            time.sleep(2)
            driver.find_element(By.ID,'show_filters').click()
#         print('show filters clicked')
        time.sleep(5)
        driver.find_element_by_id('P80_ORG_ID_BRICK_DATA').send_keys('Amgen regioner\n')
        driver.find_element_by_id('P80_ORG_ID_BRICK_DATA').send_keys(Keys.ENTER)
        time.sleep(10)
#         try:
#             driver.find_element(By.XPATH,'//*[@id="P80_WEEK_PERIOD_FROM"]/option[1]').click() # To select latest week
#             print(driver.find_element_by_xpath('//*[@id="P80_WEEK_PERIOD_FROM"]/option[1]').text)
#         except:
#             time.sleep(5)
#             driver.find_element(By.XPATH,'//*[@id="P80_WEEK_PERIOD_FROM"]/option[1]').click() # To select latest week
#             print(driver.find_element_by_xpath('//*[@id="P80_WEEK_PERIOD_FROM"]/option[1]').text)
        driver.find_element_by_id('P80_SECTORS').send_keys('Hospital Sector')
        time.sleep(5)
        
#         driver.find_element_by_id('P80_ORG_ID_BRICK_DATA').submit()
        
        
        
        try:
            timeout = 60
            locator = '/html/body/form/div[1]/div[1]/div[2]/main/div[2]/div/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div[1]/div[2]/div[2]/div[2]/div/div[2]/div[2]/div[6]/div[1]/table' # table xpath REQUIRES REVIEW
            WebDriverWait(driver, timeout).until(EC.visibility_of_element_located((By.XPATH, locator)))
        except:
            driver.refresh()
        try:
            driver.find_element(By.XPATH,'//*[@id="P80_WEEK_PERIOD_FROM"]/option[1]').click() # To select latest week
            print(driver.find_element_by_xpath('//*[@id="P80_WEEK_PERIOD_FROM"]/option[1]').text)
        except:
            time.sleep(5)
            driver.find_element(By.XPATH,'//*[@id="P80_WEEK_PERIOD_FROM"]/option[1]').click() # To select latest week
            print(driver.find_element_by_xpath('//*[@id="P80_WEEK_PERIOD_FROM"]/option[1]').text)
#         driver.find_element_by_id('P80_SECTORS').send_keys('Primary Sector')
        time.sleep(10)
        driver.find_element_by_id('data_sheet_weekly_actions_button').click()
        time.sleep(3)
        driver.find_element_by_id("data_sheet_weekly_actions_menu_14i").click()
        time.sleep(10)
        driver.find_element_by_id("data_sheet_weekly_download_CSV").click()
        #driver.find_element_by_id("data_sheet_daily_download_CSV").click()
        #driver.find_element(By.XPATH,'//*[@id="data_sheet_daily_download_CSV"]').click()
    except Exception as e:
        print(e)
        print("Error on line {}".format(sys.exc_info()[-1].tb_lineno))
        driver.close()
        driver = webdriver.Chrome(chromeDriver,options = options)
        driver.maximize_window()
        driver.get('https://dynamics.pharmaview.dk/ords/f?p=300:101')

        driver.find_element_by_id("P101_USERNAME").send_keys(username)
        driver.find_element_by_id("P101_PASSWORD").send_keys(password)
        driver.find_element_by_id('P101_LOGIN').click()
        
    retries+=1
    time.sleep(30)
    newCount = len(os.listdir(defaultDirectory))
if(retries<5):
    print("File downloaded")
    logFilePath = dfUploadParameters[(dfUploadParameters['Field'].str.lower().str.strip()=='path') & (dfUploadParameters['Source'].str.lower().str.strip()=='logfile')]['Value'].item()
    dfLogFile = pd.read_excel(logFilePath,index_col = "SNo.")
    dfLogFile.loc[dfLogFile['Script']=='PharmaviewWeeklyBrickDataHospitalSector','Executed']= 1
    dfLogFile.to_excel(logFilePath)
else:
    raise Exception("Download Error")
    #print("Download error")

driver.close()
```
